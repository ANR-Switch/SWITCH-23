```python
import pandas as pd
df = pd.read_csv("C:/Users/flavi/Travail/Switch-23/includes/population/Pop.csv",sep=";")
# Chargement de votre DataFrame
# Remplacez 'votre_dataframe.csv' par le chemin de votre fichier CSV
df = df.drop(columns='Unnamed: 0.1')
df = df.drop(columns='Unnamed: 0')

df['activities'] = df['activities'].map(lambda x : list(ast.literal_eval(x)))
df['departs'] = df['departs'].map(lambda x : list(ast.literal_eval(x)))

# Fonction pour supprimer les éléments de valeur 11 dans les listes 'activities' et 'depart'
def remove_elements(row,toDelet):
    indices_to_remove = [i for i, x in enumerate(row['activities']) if x//10 == toDelet]
    x = 0
    for i in indices_to_remove:
        row['activities'].pop(i-x)
        row['departs'].pop(i-x)
        x+=1
    

# Appliquer la fonction sur chaque ligne du DataFrame
df.loc[(df['PCSC']==3)].apply(remove_elements,args=[1], axis=1)

df.loc[(df['PCSC']==7)].apply(remove_elements,args=[2], axis=1)

df
```

```python
# Afficher le DataFrame modifié
df.to_csv("C:/Users/flavi/Travail/Switch-23/includes/population/Pop3&7TT.csv",sep=";")

df
```
