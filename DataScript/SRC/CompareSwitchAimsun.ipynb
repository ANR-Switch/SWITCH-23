```python

trajet = switch_road.groupby(['ID']).agg({
                                    "distance": lambda x: sum(x),\
                                    "TopoId" : lambda x: list(np.unique(x)),\
                                    "duration": lambda x: sum(x)
})
trajet
```

```python

switch_road_ = switch_road.groupby(["TopoID"],group_keys=True).agg({
                                        "distance": lambda x: list(np.unique(x))[0],\
                                        "duration": lambda x: np.mean(x),\
                                        "meanSpead":lambda x: np.mean(x),\
                                        "VehicleID" : lambda x:len(list(np.unique(x)))
})

switch_road_2 = switch_road_.rename(columns ={"VehicleID":"frequentation",'duration':'meanDuration'})
switch_road_2

```
